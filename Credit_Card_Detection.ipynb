```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Create dataset
X = np.array([[2000, 8], [5000, 2], [3000, 12]])
y = [1, -1, 1]  # 1 = Valid, -1 = Fraudulent

# Create and train the SVM model
model = svm.SVC(kernel='linear')
model.fit(X, y)

# New data point
new_data = np.array([[4000, 1]])
prediction = model.predict(new_data)

print("Validity of the new data point:", "Valid" if prediction[0] == 1 else "Fraudulent")

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=['blue' if label == 1 else 'red' for label in y], s=50)
plt.scatter(new_data[:, 0], new_data[:, 1], c='green', s=100, marker='x')
plt.xlabel('Amount ($)')
plt.ylabel('Time (hours)')
plt.title('SVM Credit Card Fraud Detection')
plt.show()
```
