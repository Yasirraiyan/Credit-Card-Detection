```python
import numpy as np
from cvxopt import matrix, solvers

# Define the parameters for the QP problem
P = matrix(np.array([[68.0, -26.0, 102.0], [-26.0, 29.0, -39.0], [102.0, -39.0, 153.0]]), tc='d')
q = matrix(np.array([-1.0, -1.0, -1.0]), tc='d')
G = matrix(np.diag([-1.0, -1.0, -1.0]))
h = matrix(np.zeros(3))
A = matrix(np.array([[1.0, -1.0, 1.0]]), (1, 3), 'd')
b = matrix(np.array([0.0]))

# Solve the QP problem
solution = solvers.qp(P, q, G, h, A, b)
alphas = np.array(solution['x'])

print("Optimal values of alpha:", alphas)
```
